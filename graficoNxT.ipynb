```python
import matplotlib.pyplot as plt

# Dados fornecidos
N = [100_000_000, 200_000_000, 300_000_000, 800_000_000]
T = [0.719, 0.729, 0.730, 0.735]

# Criar gr√°fico
plt.figure(figsize=(10,5))
plt.plot(N, T, marker='o', linewidth=2)
plt.xlabel("N (tamanho do vetor)", fontsize=12)
plt.ylabel("Tempo (s)", fontsize=12)
plt.title("Tempo x Tamanho do Vetor", fontsize=14)
plt.grid(True)

plt.show()
```
